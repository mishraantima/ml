```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
a=["ami","antima","krisha","kajal"]
b=[1,2,3,4]
p=["r","y","b","g"]
plt.bar(a,b,width=0.4,color=p,linestyle=":",edgecolor="y",linewidth=7)
plt.title("Bar chart",color="b")

plt.show()
```

```python
x=np.linspace(0,100,10)
y=x*x
```

```python
print("x= :",x)
print("y=: ",y)
plt.plot(x,y)
plt.title("Line chart",color="r")
plt.xlabel("value")
plt.ylabel("category")

```

```python
a=["ami"]
b=[1]
plt.bar(a,b,width=0.4,color="r")
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
category=np.array(["a","b","c","d"])
v1=np.array([1,3,5,6])
v2=np.array([2,4,6,8])
v3=np.array([11,12,13,14])
plt.bar(category,v1,label="set1")
plt.bar(category,v2,label="set2")
plt.bar(category,v3,label="set3")
plt.show()
```

```python
category=np.array(["ami","kajal","anu","krisha"])
v1=np.array([1,3,5,6])


plt.pie(v1,labels=category,autopct="%1.1f%%")

plt.show()
```

```python
categories=np.array(["ami","kajal","anu","krisha"])
v1=np.array([1,3,5,6])
fig, ax=plt.subplots()
ax.pie(v1, labels=categories, radius=1,startangle=90,counterclock=False)
ax.pie([1],radius=0.6,colors="w")
ax.set_title("Donut chart")
plt.show()

```
